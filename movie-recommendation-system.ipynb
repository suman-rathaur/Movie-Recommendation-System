# Feature Selection

Creating a new dataframe called 'movies which consists of only essential features

```python
movies = df[['id','genres' ,'title', 'overview']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
#Creating a new column
movies['tags'] = movies['overview'] + movies['genres']
```

```python
#Deleting overview & genres
movies = movies.drop(['overview', 'genres'],axis="columns")
```

```python
movies.shape
```

We will use below models from sklearn library:
1) Bag of Words
2) TFIDF models 

# Bag of Words Model

## Creating an instance of CountVectorizer Class

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#we limited to 4803 because that is the shape of movies
cv = CountVectorizer(max_features=4803, stop_words="english")
```

## Calling fit() function, creating document vectors

```python
#Converting data to unicode
vector = cv.fit_transform(movies['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

# Cosine Similairty

Cosine similarity measures the similarity between two vectors of an inner product space. It is measured by the cosine of the angle between two vectors and determines whether two vectors are pointing in roughly the same direction. It is often used to measure document similarity in text analysis.

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity
```

# Indexing

```python
#Indexing
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(movies.iloc[i[0]].title)
```

# Function to Recommend movie

```python
def recommend(mov):
    index=movies[movies['title']==mov].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(movies.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```
